```python
#Necessary libraries for Eigen Face implementation
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
```

```python
#Step-1: read all images from the data set and convert it into a FP matrix
image_list = []
for file in glob.glob('large_set/*.jpg'):
    im = cv2.imread(file)
    im = np.float32(im)/255.0
    image_list.append(im)
print("Each image size is: ", image_list[0].shape)
```

```python
#Step-2: Convert the MXNX3 matrix into a 1D matrix considering each pixel (R,G,B) as an independent variable.
data = []
for img in image_list:
    data.append(img.flatten())
data = np.array(data)
print("Each image size in 1-D: ",data[0].shape[0])
```

```python
#Step-3: Compute PCA using opencv2 library function and display the average face
mean, eigvector = cv2.PCACompute(data, mean=None, maxComponents=10)
avgface = mean.reshape(image_list[0].shape)
plt.title("Mean Face")
plt.imshow(avgface)
```

```python
#Step-4: Resize all eigen vectors
eigenFaces = []
for vec in eigvector:
    eigenFaces.append(vec.reshape(image_list[0].shape))
```

```python
#Display the major component variations of images
fig = plt.figure()
plt.title('Principal Component PC-1 variation')
fig.add_subplot(1,2,1)
PC1op1 = np.add(avgface, eigenFaces[0]*-90)
PC1op2 = np.add(avgface, eigenFaces[0]*100)
plt.imshow(PC1op1)
fig.add_subplot(1,2,2)
plt.imshow(PC1op2)
```
